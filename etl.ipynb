```python
import os
import json
import pyodbc

# SQL connection
conn = pyodbc.connect(
    "Driver={ODBC Driver 17 for SQL Server};"
    "Server=localhost\\sqlexpress;"
    "Database=phonepe;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()
cursor.fast_executemany = True

# Path
base_path = r"D:\LABMENTIX\phone_pe\pulse-master\data\aggregated\transaction\country\india\state"

for state in os.listdir(base_path):
    state_path = os.path.join(base_path, state)

    if not os.path.isdir(state_path):
        continue

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            if not file.endswith(".json"):
                continue

            file_path = os.path.join(year_path, file)

            with open(file_path, 'r') as f:
                data = json.load(f)

                if data.get("data") is None:
                    continue

                quarter = file.replace(".json", "")

                for item in data["data"].get("transactionData", []):
                    t_type = item.get("name")

                    for pay in item.get("paymentInstruments", []):
                        cursor.execute("""
                            INSERT INTO dbo.aggregated_transaction
                            (state, year, quarter, transaction_type, transaction_count, transaction_amount)
                            VALUES (?, ?, ?, ?, ?, ?)
                        """,
                        state,
                        int(year),
                        int(quarter),
                        t_type,
                        int(pay.get("count")),
                        float(pay.get("amount"))
                        )

conn.commit()
cursor.close()
conn.close()

print("✅ aggregated_transaction inserted successfully!")
```
