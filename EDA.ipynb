## 📊 SQL Data Analysis

The following SQL queries were executed to extract insights from the data.

```python
query = """
SELECT ROUND(SUM(transaction_amount),2) AS total_amount
FROM aggregated_transaction
"""
pd.read_sql(query, conn)
```

### 🔍 Insight:

- Total transaction value is very high, indicating strong digital adoption.
- PhonePe has a large transaction volume across India.

```python
query = """
SELECT TOP 5 state, SUM(transaction_amount) AS total
FROM aggregated_transaction
GROUP BY state
ORDER BY total DESC
"""
pd.read_sql(query, conn)
```

### 🔍 Insight:

- A few states dominate transaction volume.
- These regions are key revenue drivers.

```python
query = """
SELECT transaction_type, SUM(transaction_amount) AS total
FROM aggregated_transaction
GROUP BY transaction_type
ORDER BY total DESC
"""
pd.read_sql(query, conn)
```

### 🔍 Insight:

- Recharge and peer-to-peer payments are the most common.
- Indicates frequent daily usage of the platform.

```python
query = """
SELECT year, SUM(transaction_amount) AS total
FROM aggregated_transaction
GROUP BY year
ORDER BY year
"""
pd.read_sql(query, conn)
```

### 🔍 Insight:

- Transactions are increasing year by year.
- Shows continuous growth in digital payments.

```python
query = """
SELECT TOP 10 district, SUM(transaction_amount) AS total
FROM top_district
GROUP BY district
ORDER BY total DESC
"""
pd.read_sql(query, conn)
```

### 🔍 Insight:

- Certain districts show very high activity.
- Useful for targeted marketing campaigns.

```python
query = """
SELECT TOP 10 pincode, SUM(transaction_amount) AS total
FROM top_pincode
GROUP BY pincode
ORDER BY total DESC
"""
pd.read_sql(query, conn)
```

### 🔍 Insight:

- High activity at pincode level shows localized usage.
- Enables hyperlocal targeting strategies.

```python
import matplotlib.pyplot as plt

query = """
SELECT TOP 5 state, SUM(transaction_amount) AS total
FROM aggregated_transaction
GROUP BY state
ORDER BY total DESC
"""

df = pd.read_sql(query, conn)

df.plot(kind='bar', x='state', y='total')
plt.title("Top 5 States by Transaction")
plt.show()
```

```python
df.plot(kind='bar', x='state', y='total', title='Top States')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

plt.pie(
    df['total'],
    labels=None,              # ❗ labels remove from pie
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.7           # percentage inside adjust
)

# ✅ Legend outside (top-right corner)
plt.legend(
    df['transaction_type'],
    title="Transaction Type",
    loc="upper right",
    bbox_to_anchor=(1.3, 1)   # position adjust
)

plt.title("Transaction Type Distribution")
plt.tight_layout()
plt.show()
```

```python
df_year = pd.read_sql("""
SELECT year, SUM(transaction_amount) as total
FROM aggregated_transaction
GROUP BY year
ORDER BY year
""", conn)

df_year.plot(kind='line', x='year', y='total', title='Yearly Growth')
```

```python
df_district = pd.read_sql("""
SELECT TOP 10 district, SUM(transaction_amount) as total
FROM top_district
GROUP BY district
ORDER BY total DESC
""", conn)

df_district.plot(kind='bar', x='district', y='total', title='Top Districts')
```

```python
df_pincode = pd.read_sql("""
SELECT TOP 10 pincode, SUM(transaction_amount) as total
FROM top_pincode
GROUP BY pincode
ORDER BY total DESC
""", conn)

df_pincode.plot(kind='bar', x='pincode', y='total', title='Top Pincodes')
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px
from PIL import Image
import os

# =====================
# PAGE CONFIG
# =====================
st.set_page_config(layout="wide")

# =====================
# LOAD LOGO
# =====================
logo_path = r"D:\LABMENTIX\phone_pe\logo.png"

logo = None
if os.path.exists(logo_path):
    logo = Image.open(logo_path)

# =====================
# LOAD DATA
# =====================
df = pd.read_csv("top_pincode.csv")

# =====================
# SIDEBAR
# =====================
st.sidebar.title("🔍 Filters")

state_list = ["All"] + sorted(df["state"].dropna().unique())
year_list = sorted(df["year"].dropna().unique())
quarter_list = sorted(df["quarter"].dropna().unique())

state = st.sidebar.selectbox("Select State", state_list)
year = st.sidebar.selectbox("Select Year", year_list)
quarter = st.sidebar.selectbox("Select Quarter", quarter_list)

# Footer (NO LOGO)
st.sidebar.markdown("""
<style>
/* Sidebar ko full height flex bana */
section[data-testid="stSidebar"] > div:first-child {
    display: flex;
    flex-direction: column;
    height: 100vh;
}

/* Footer bottom push */
.sidebar-footer {
    margin-top: auto;
    text-align: center;
    padding-bottom: 20px;
}

.sidebar-footer p {
    margin: 0;
}
</style>

<div class="sidebar-footer">
    <p style='color:white;'>👤 Developed by</p>
    <p style='color:yellow;font-weight:bold;'>Rakesh Mahakur</p>
</div>
""", unsafe_allow_html=True)

# =====================
# FILTER LOGIC
# =====================
filtered = df.copy()

if state != "All":
    filtered = filtered[filtered["state"] == state]

filtered = filtered[filtered["year"] == year]
filtered = filtered[filtered["quarter"] == quarter]

# =====================
# CSS
# =====================
st.markdown("""
<style>
.stApp {background-color: #0b1220;}

section[data-testid="stSidebar"] {
    background: linear-gradient(180deg,#5f259f,#7b3fe4);
}

section[data-testid="stSidebar"] * {
    color:white !important;
}

.kpi-card {
    background:#dcd6e8;
    padding:20px;
    border-radius:15px;
    text-align:center;
    color:black !important;
}

/* Chart card */
[data-testid="stPlotlyChart"] {
    background:white;
    padding:10px;
    border-radius:12px;
}

/* Remove spacing issue */
.block-container {
    padding-left:2rem;
    padding-right:2rem;
}

.js-plotly-plot text {
    fill:black !important;
}
</style>
""", unsafe_allow_html=True)

st.markdown("""
<style>

/* WHITE VERTICAL DIVIDER */
section[data-testid="stSidebar"] {
    border-right: 4px solid white;  /* 👈 yeh line add kar */
}

/* Optional: smooth look */
section[data-testid="stSidebar"] {
    box-shadow: 2px 0 10px rgba(255,255,255,0.2);
}

</style>
""", unsafe_allow_html=True)

# =====================
# HEADER (NO LOGO)
# =====================
st.markdown("""
<h1 style='text-align:center; font-size:40px; font-weight:700;'>
PhonePe Advanced Dashboard
</h1>
""", unsafe_allow_html=True)

# =====================
# KPI
# =====================
total = filtered["amount"].sum()
states = filtered["state"].nunique()

col1, col2 = st.columns(2)

with col1:
    st.markdown(f"""
    <div class='kpi-card'>
        <b>💰 Total Transaction</b><br><br>
        <h2>{total:,.0f}</h2>
    </div>
    """, unsafe_allow_html=True)

with col2:
    st.markdown(f"""
    <div class='kpi-card'>
        <b>📍 Total States</b><br><br>
        <h2>{states}</h2>
    </div>
    """, unsafe_allow_html=True)

# =====================
# PIE CHART (SMALL + CENTER)
# =====================
st.subheader("🟣 State Contribution")

pie = filtered.groupby("state")["amount"].sum().reset_index()

fig_pie = px.pie(
    pie,
    values="amount",
    names="state",
    hole=0.6,
    width=500,
    height=500
)

fig_pie.update_traces(textinfo='percent+label')

fig_pie.update_layout(
    paper_bgcolor="white",
    plot_bgcolor="white",
    font=dict(color="black"),
    margin=dict(l=10, r=10, t=10, b=10)
)

col1, col2, col3 = st.columns([2,3,2])

with col2:
    st.plotly_chart(fig_pie)

# =====================
# BAR + PINCODE
# =====================
col1, col2 = st.columns(2)

with col1:
    st.subheader("🏆 Top States")
    top_states = filtered.groupby("state")["amount"].sum().nlargest(10).reset_index()

    fig_bar = px.bar(top_states, x="state", y="amount", color="amount")

    fig_bar.update_layout(
        paper_bgcolor="white",
        plot_bgcolor="white",
        font=dict(color="black")
    )

    st.plotly_chart(fig_bar, use_container_width=True)

with col2:
    st.subheader("📍 Top Pincodes")
    top_pin = filtered.nlargest(10, "amount")

    fig_pin = px.bar(top_pin, x="pincode", y="amount", color="amount")

    fig_pin.update_layout(
        paper_bgcolor="white",
        plot_bgcolor="white",
        font=dict(color="black")
    )

    st.plotly_chart(fig_pin, use_container_width=True)

# =====================
# LINE CHART
# =====================
st.subheader("📈 Year Trend")

year_data = filtered.groupby("year")["amount"].sum().reset_index()

fig_line = px.line(year_data, x="year", y="amount")

fig_line.update_layout(
    paper_bgcolor="white",
    plot_bgcolor="white",
    font=dict(color="black")
)

st.plotly_chart(fig_line, use_container_width=True)
```
